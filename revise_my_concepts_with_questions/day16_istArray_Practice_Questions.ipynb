```python
def find_max_element(lst):
  max_element = lst[0]
  for num in lst:
        if num > max_element:
            max_element = num
  return max_element


def display_result(lst):
    print(find_max_element(lst))

lst = [3,4,2,4,23,4,243,2,3234,234,23,4234,23,423,435,4,5,67,567,78,687,6]
display_result(lst)
```

```python
def sum_of_elements(lst):

    total_sum = 0

    for num in lst:
        total_sum += num

    return total_sum


def display_result(lst):
    print(sum_of_elements(lst))


lst = [1, 2, 3, 4, 5]
display_result(lst)
```

```python
def is_palindrome(lst):

    left = 0
    right = len(lst) - 1

    while left < right:
        if lst[left] != lst[right]:
            return False
        left += 1
        right -= 1

    return True

def display_result(lst):
    print(is_palindrome(lst))


lst = [1, 2, 3, 2, 1]
display_result(lst)
```

```python
def reverse_list(lst):

    left = 0
    right = len(lst) - 1

    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1

    return lst

def display_result(lst):
    print(reverse_list(lst))

lst = [1, 2, 3, 4, 5]
display_result(lst)
```

```python
def rotate_left(arr, k):

    k = k % len(arr)
    return arr[k:] + arr[:k]


def display_result(arr, k):
    print(rotate_left(arr, k))

arr = [1, 2, 3, 4, 5]
k = 2
display_result(arr, k)
```

```python
def plus_one(digits):


    n = len(digits)
    for i in range(n - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        digits[i] = 0


    return [1] + digits

def display_result(digits):
    print(plus_one(digits))


digits = [9, 5, 9]
display_result(digits)
```

```python
def find_missing_number(nums):

    n = len(nums)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum

def display_result(nums):
    print(find_missing_number(nums))

nums = [3, 0, 1]
display_result(nums)
```

```python
def is_sorted(arr):
    n = len(arr)

    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            return False

    return True

def display_result(arr):
    print(is_sorted(arr))


arr = [1, 2, 3, 4, 5]
display_result(arr)
```

```python
def intersection(nums1, nums2):


    set1 = set(nums1)
    intersection_set = set()

    for num in nums2:
        if num in set1:
            intersection_set.add(num)
    return list(intersection_set)

def display_result(nums1, nums2):
    result = intersection(nums1, nums2)
    print(result)


nums1 = [4, 9, 5]
nums2 = [9, 4, 9, 8, 4]
display_result(nums1, nums2)
```
