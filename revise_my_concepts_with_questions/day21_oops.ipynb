```python
# modeling a bank account

#  defines a class for bank account

class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    print(f"{amount} is depsoited. new balance is {self.balance}")

  def withdraw(self, amount):
    if amount>self.balance:
      print("Insufficent funds!")
    else:
      self.balance -= amount
      print(f"{amount} is withdraw. New Balance is {self.balance}")

  def get_balance(self):
    return self.balance

#  create an account

account = BankAccount("krish", 80090)
print(account.balance)



account.deposit(1999)
```

```python
#  inherit class for tesla car

class Tesla(car):
  def __init__(self, windows, doors, enginetype, is_self_drive):
    super().__init__(windows, doors, enginetype)
    self.is_self_drive = is_self_drive


  def self_drive(self):
    print(f"Tesla supports self driving: {self.is_self_drive}")


tesla1 = Tesla(3,5,"giant", "yes")
tesla1.self_drive()
```

```python
# multiple inheritance
#  when a class inherits from more than one base class

# base class

class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    print("Subclass must implement this method ")

# base class 2
class Pet:
  def __init__(self, owner):
    self.owner = owner

# derived class
class dog(Animal, Pet):
  def __init__(self, name, owner):
    Animal.__init__(self,name)
    Pet.__init__(self,owner)

  def speak(self):
    return f"{self.name} say woof"


#  create an object

Dog = dog("buddy", "krish")
print(Dog.speak())
print(f"owner:{Dog.owner} ")
```

```python
# abstraction

from abc import ABC, abstractmethod

## abstract base class

class Vehicle(ABC):
  def drive(self):
    print("The vehicle is used for driving")

  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  def start_engine(self):
    print("car enginer started")

def operate_vehicle(vehicle):
  vehicle.start_engine()
  vehicle.drive()

car = Car()
operate_vehicle(car)
```

```python
# magic methods

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name}, {self.age} years old"


  def __repr__(self):
    return f"Person(name={self.name}, age = {self.age})"

person = Person("hello", 87)
print(person)
```
