```python
# read whole file

with open('source.txt', 'r') as file:
  content = file.read()
  print(content)

```

```python

with open('source.txt', 'r') as file:
  for line in file:
    print(line.strip())
```

```python
# overide the previous file
with open("source.txt", "w") as file:
  file.write("hello")
  file.write("hello sodhfoia ")
  file.write("hellshfa  dhioafhd apo")
  file.write("he doidi ja opdj f ad j90llo")
```

```python
with open("source.txt", "a") as file:
  file.write("hell  dk d ie id k o")
```

```python
lines = ["first line \n second file \n third file"]
```

```python
with open("source.txt", "a") as file:
  file.writelines(lines)
```

```python
# reading binary files

with open('example.bin', 'rb') as file:
  content = file.read()
  print(content)
```

```python
with open('source.txt', 'w+') as file:
  file.write("hello world \n")
  file.write("this is a new line of character \n")

  # move the file cursor to the begineeing
  file.seek(0)

  print(content)
```
